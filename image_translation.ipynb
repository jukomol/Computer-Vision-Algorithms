# Understanding Translation Concept

Translation is a transformation that moves every pixel of an image by a fixed distance in a given direction.

If an image is translated by \((t_x, t_y)\), then each pixel at \((x, y)\) is moved to a new location:

\[
(x', y') = (x + t_x, y + t_y)
\]

where:
- \(t_x\) is the translation along the horizontal axis (left/right).
- \(t_y\) is the translation along the vertical axis (up/down).

## Representation Using Transformation Matrix

Translation can be expressed as a matrix operation:

\[
\begin{bmatrix}
x' \\
y' \\
1
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & t_x \\
0 & 1 & t_y \\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
1
\end{bmatrix}
\]

This homogeneous transformation matrix makes it easier to apply translation in computer vision applications.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Image Translation\IMG_9422.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.axis('off')
```

```python
tx, ty = 50,30 #shift 50 pixels in x direction and 30 pixels in y direction
translation_matrix = np.float32([[1,0,tx],[0,1,ty]]) #translation matrix for shifting image by tx and ty pixels
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0])) #applying translation matrix to image using warpAffine function of opencv library
```

```python
#plt.subplot(1,2,1)
plt.imshow(translated_image)
plt.axis('off')
plt.show()
```
